```python
import pandas as pd
datos = pd.read_csv("celsius.csv")
```

```python
datos.info()
```

```python
datos.head()
```

```python
import seaborn as sb

sb.scatterplot(data=datos,x="celsius",y=" fahrenheit",hue=" fahrenheit",palette="coolwarm")
```

```python
# Características (X), etiqueta(y)
x = datos["celsius"]
y = datos[" fahrenheit"]
```

```python
y
```

```python
#x.values
#[-40, -10, 0, 8, ...]
# Buscamos esto:
#[[-40], [-10], [0], [8], ...[38]]

# y para lograrlo utilizamos:
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
```

```python
# Entrenamiento
modelo.fit(x_procesada,y_procesada)
```

```python
celsius = 111
prediccion = modelo.predict([[celsius]])
print(f"{celsius} grados celsius son {prediccion} grados fahrenheit")
```

```python
modelo.score(x_procesada,y_procesada)
# Esto devuelve que tan bien está entrenada la ia, mientras mas cercana a uno mejor.
```

```python
# ¿Qué es la regresión lineal y para qué sirve?
# 
# Nos sirve para modelar la relación entre la llamada variable dependiente
# (respuestas) y una o mas independientes (entradas)
# 
# y = mx+b
# 
# La linea que mejor se adapte a los datos toma en cuenta a la menor suma
# de los residuos al cuadrado, generando ahi el linea/plano/etc#
```
