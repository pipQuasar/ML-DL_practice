```python
import pandas as pd # type: ignore
datos = pd.read_csv("exercise_03.csv")
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(data=datos,x="centimetros",y="metros")
```

```python
x = datos["centimetros"]
y = datos["metros"]
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(x_procesada,y_procesada)
```

```python
centimetros = 4736
prediccion = modelo.predict([[centimetros]])
print(f"{centimetros} centimetros es igual a {prediccion} metros!")
```
