```python
import pandas as pd # type: ignore
datos = pd.read_csv("exercise_01.csv")
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(data=datos,x="kilometros",y="millas")
```

```python
x = datos["kilometros"]
y = datos["millas"]
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(x_procesada,y_procesada)
```

```python
kilometros = 66
prediccion = modelo.predict([[kilometros]])
print(f"{kilometros} Kilometros son aproximadamente {prediccion} millas!") 
```
