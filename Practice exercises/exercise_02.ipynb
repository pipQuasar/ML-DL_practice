```python
import pandas as pd #type: ignore
datos = pd.read_csv("exercise_02.csv")
```

```python
datos.head()
```

```python
import seaborn as sb
sb.scatterplot(data=datos,x="metros",y="pies")
```

```python
x = datos["metros"]
y = datos["pies"]
# Debo separar cada uno de sus valores en arrays unicos.
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
# Lo entreno
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(x_procesada,y_procesada)
```

```python
metros = 100
prediccion = modelo.predict([[metros]])
print(f"{metros} metros son {prediccion} pies!")
```
